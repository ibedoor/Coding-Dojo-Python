Import Laibraries

```
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

Upload Dataset

```
df = sns.load_dataset('titanic')
df
```

###**Visualization**

Heatmap for Dataset

```
sns.heatmap(df.corr(), annot=True)
plt.title("Distribution of Survival", fontsize =15);
```

Distribution Numbers of People Who Survived and Who Died

```
plt.figure(figsize=(12,6)) 
SV = df.survived.value_counts()
SV.index = ['Died','Survived']
SV.plot(kind='bar', color='thistle')
plt.title("Distribution of Survival", fontsize =15);
```

Which The Class Have the Hieghest Fare Amount?

```
plt.figure(figsize=(12, 8)) 
sns.lineplot(x='class',  y='fare', alpha=0.5, data= df, estimator=None, color='thistle')
plt.title("Distribution of Classes with Fare Amount",  fontsize =15);
```

Which Number of Age Groups Was the Highest?

```
people = df.who.value_counts()
plt.figure(figsize=(12,6)) 
people.index = ['Children','Women', 'Man']
people.plot(kind='bar', color='thistle')
plt.title("Distribution of Age Group", fontsize =15);
```
